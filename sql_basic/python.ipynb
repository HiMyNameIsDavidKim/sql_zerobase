```python
mydb = mysql.connector.connect(
            host = 'localhost',
            user = 'root',
            password = 'root',
            database = 'zerobase',
        )

cur = mydb.cursor()
sql = open('test04.sql').read()
for i in cur.execute(sql, multi=True):
    if i.with_rows:
        print(i.fetchall())
    else:
        print(i.statement)

mydb.commit()
mydb.close()
```

```python
mydb = mysql.connector.connect(
            host = 'localhost',
            user = 'root',
            password = 'root',
            database = 'zerobase',
        )

cur = mydb.cursor(buffered=True)
cur.execute('select * from sql_file')

result = cur.fetchall()
for i in result:
    print(i)

mydb.close()
```
